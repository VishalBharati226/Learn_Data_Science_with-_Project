### Question 1: Print 1 to N

Write a program that takes an integer n as input and prints numbers from 1 to n.

```python
n = int(input("Enter number here:"))
for i in range(1,n+1):
    print(i)
```

### Question 2: Sum of First N Numbers

Write a program to calculate the sum of first n natural numbers.

```python
n = int(input("Enter number here:"))

sum =0
for i in range(1,n+1):
    print(i)
    sum= sum+i
print("total sum of n natural number: ",sum)
```

### Question 4: Factorial of a Number

Write a program to calculate the factorial of a number n.

```python
num = int(input("Enter number here: "))
fact = 1
for i in range(1,num+1):
    fact = fact*i

print(fact)

```

### Question 5: Reverse a Number

Write a program to reverse a given number.

```python
num = int(input("Enter number here: "))

rev = 0

while num > 0:
    digit = num % 10
    rev = rev * 10 + digit
    num = num // 10

print(rev)
```

### Question 6: Count Digits in a Number

Write a program to count the number of digits in a given number.

```python
num = int(input("Enter number here: "))

count = 0

while num > 0:
    num = num // 10
    count += 1

print(count)
```

### Question 7: Check Prime Number

Write a program to check whether a given number is prime or not.

```python
num = int(input("Enter number here: "))
if num <=1:
    print("num is prime is prime")
else:
    for i in range(2,num):
        if num % i== 0:
            print("num is prime")
            break
    else:
        print("num is not prime")
```

### Question: Find Leap Year

Write a program to check whether a given year is a leap year or not.

```python
start = int(input("Enter start year: "))
end = int(input("Enter end year: "))

for year in range(start, end + 1):
    if year % 400 == 0:
        print(year, "leap year")
    elif year % 100 == 0:
        print(year, "not leap year")
    elif year % 4 == 0:
        print(year, "leap year")
    else:
        print(year, "not leap year")
```
