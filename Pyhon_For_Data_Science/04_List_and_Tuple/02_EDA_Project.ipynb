## Overview:

This project focuses on analyzing weekly sales data of a store using Python basics like loops, conditions, and lists.

The goal is to think like a data analyst and understand how raw numbers convert into business insights.

### Problem Statement:

A store has provided daily sales data for a week. You need to analyze this data using basic Python concepts.

```python
sales = [1200, 3400, 2800, 1500, 5000, 2300, 4100]
days = ["Mon", "Tue", "Wed", "Thu", "Fri", "Sat", "Sun"]
```

### Problem 1: Calculate the total sales for the week.

```python
total_sales = 0

for i in sales:
    total_sales += i

print(total_sales)
```

### Problem 2: Find the average sales for the week.

Calculate the average of all daily sales by dividing the total sales by the number of days.

```python
days_count = len(days)
average_sales = total_sales / days_count

print(average_sales)
```

### Problem 3: Find the highest sales value and the corresponding day.

Find the maximum sales value from the list and also print the day on which it occurred.

```python
highest_sale = max(sales)

index = sales.index(highest_sale)
highest_day = days[index]

print(highest_sale)
print(highest_day)
```

### Problem 4: Find the lowest sales value and the corresponding day.

Find the minimum sales value from the list and also print the day on which it occurred.

```python
lowest_sale = min(sales)
index = sales.index(lowest_sale)
lowest_day = days[index]
print(lowest_sale)
print(lowest_day)
```

### Problem 5: Count the number of days where sales were greater than 3000.

Count how many days in the list had sales strictly greater than 3000.

```python
count = 0

for i in sales:
    if i > 3000:
        count += 1

print(count)
```

### Problem 6: Print all days where sales were less than 2000.

Print all the days corresponding to sales values that are strictly less than 2000.

```python
for i in range(len(sales)):
    if sales[i] < 2000:
        print(days[i])
```

### Problem 7: Find the difference between highest and lowest sales.

Calculate the difference between the maximum sales value and the minimum sales value in the list.

```python
difference_of_sale = highest_sale-lowest_sale
print(difference_of_sale)
```

### Problem 8: Print day-wise sales in proper format.

Print each day along with its corresponding sales value in the format:

```python
for i in range(len(sales)):
    print(days[i], "->", sales[i])
```

### Observations:
. Sales data shows fluctuation across all days of the week

. Some days have significantly higher sales compared to others

. A few low-performing days are also present

. Majority of revenue is contributed by high-sales days

. No uniform pattern is observed in daily sales

### Insights:
. Peak sales day significantly boosts total weekly revenue

. More than half of total sales come from just 2–3 days

. Low sales days are pulling down overall average performance

. Business is dependent on a few strong-performing days

. If weak days are improved, overall revenue can increase noticeably

### Conclusion:

The sales performance is inconsistent throughout the week. A few peak days contribute most of the revenue, while some days show weak performance. Overall, the business has good potential but lacks consistency in daily sales.
