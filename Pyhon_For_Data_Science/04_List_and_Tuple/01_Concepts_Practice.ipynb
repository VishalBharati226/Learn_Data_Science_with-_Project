### Question 4: Largest Element in List

Write a program to find the largest number in a given list without using max().

```python
nums = [10, 5, 20, 8, 25]
largest = nums[0]
for i in nums:
    if i > largest:
        largest = i
print(largest)
```

### Question 5: Reverse a List

Write a program to reverse a given list without using built-in methods like reverse() or slicing ([::-1]).

```python
nums = [1, 2, 3, 4, 5]

reversed_list = []

for i in range(len(nums) - 1, -1, -1):
    reversed_list.append(nums[i])

print(reversed_list)
```

### Question 6: Count Occurrences in List

Write a program to count how many times a given number appears in a list without using built-in count() function.

```python
nums = [1, 2, 3, 2, 4, 2, 5]
target = 2
occ = 0
for i in nums:
    if i == target:
        occ +=1
print(occ)
```

### Question 7: Remove Duplicates from List

Write a program to remove duplicate elements from a list without using set().

```python
nums = [1, 2, 2, 3, 4, 4, 5]

unique = []

for i in nums:
    if i not in unique:
        unique.append(i)

print(unique)
```

### Question 8: Sum of First N Natural Numbers

Write a program to find the sum of first n natural numbers without using formulas or built-in functions.

```python
n = int(input("enter number here:"))

total = 0

for i in range(1, n + 1):
    total += i

print(total)
```

### Question 10: Count Even and Odd Numbers

Write a program to count how many even and odd numbers are present in a given list.

```python
nums = [1, 2, 3, 4, 5, 6, 7, 8]

even = 0
odd = 0

for i in nums:
    if i % 2 == 0:
        even += 1
    else:
        odd += 1

print("Even:", even)
print("Odd:", odd)
```
