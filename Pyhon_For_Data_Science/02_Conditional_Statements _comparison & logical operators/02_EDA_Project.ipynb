# ⚡ Electricity Bill Analysis Project

## Objective

The goal of this project is to calculate electricity bills based on units consumed and perform basic data analysis.

We will:
- Take user input for electricity usage
- Apply slab-based billing logic
- Store multiple users' data
- Perform basic analysis like average, max, and min bill

## Billing Problem to solve if 

- First 100 units → ₹5 per unit  
- Next 100 units (101–200) → ₹7 per unit  
- Above 200 units → ₹10 per unit  

This is called **slab-based billing**, which is used in real-world electricity systems.

## Python code here

```python
unit = int(input("Enter electricity units consumed: "))
if unit < 0:
    print("Invalid electricity unit")
elif 0 <= unit <= 100:
    bill = unit * 5 
elif 101<= unit <= 200:
    bill = (100 * 5)+(unit-100)*7
else:
    bill = (100*5)+(100*7)+(unit-200)*10

print("Electricity unit consumed is:", unit)
print("Total Electricity Bill is: ₹",bill)
```

## Observation

- Bill increases as units increase
- Higher usage leads to higher cost per unit
