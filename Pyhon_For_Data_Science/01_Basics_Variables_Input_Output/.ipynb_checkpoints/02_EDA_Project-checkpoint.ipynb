### ❓ Q6. Simple Interest

```python
p = float(input("Enter Principal: "))
r = float(input("Enter Rate: "))
t = float(input("Enter Time: "))

si = (p * r * t) / 100

print(f"Simple Interest is: {si}")
```

### ❓ Q9. Multiple Inputs in One Line

```python
data = input("Enter two numbers: ")
a, b = data.split()

a = int(a)
b = int(b)

print(f"Sum is {a + b}")
```

```python
data = input("Enter the three input here : ")
a,b,c = data.split()
a =int(a)
b = int(b)
c = int(c)
print(f"total sum is { a+b+c}")
```

### ❓ Q3: Average of 3 Numbers

```python
data = input("Enter 3 numbers: ")
a, b, c = data.split()

a = int(a)
b = int(b)
c = int(c)

average = (a + b + c) / 3
print(f"Average is {average}")
```
