# Student Performance Analysis (Using Variables, Data Types, print(), input())

**Objective:**  
- Collect student data using `input()`
- Store data in variables with correct data types (`str`, `int`, `float`, `bool`)
- Perform basic calculations (total, average)
- Print summary using `print()`

```python
# Collect student basic info
student_name = input("Enter student name: ")
age = int(input("Enter age: "))       # integer
gender = input("Enter gender (M/F): ") 

# Collect marks for subjects
marks_math = float(input("Enter Math marks: "))  
marks_science = float(input("Enter Science marks: "))  
marks_english = float(input("Enter English marks: "))  
```

```python
# Calculate total and average marks
total_marks = marks_math + marks_science + marks_english
average_marks = total_marks / 3

# Determine pass/fail using boolean
pass_status = average_marks >= 40  # True if average >= 40
```

```python
print("\n----- Student Performance Summary -----")
print("Name:", student_name)
print("Age:", age, "years")
print("Gender:", gender)
print("Math:", marks_math, "| Science:", marks_science, "| English:", marks_english)
print("Total Marks:", total_marks)
print("Average Marks:", round(average_marks, 2))
print("Pass Status:", "Pass" if pass_status else "Fail")
```
